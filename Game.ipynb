```python
from tkinter import *
from PIL import Image, ImageTk

root = Tk()
root.title("Tic-Tac-Toe Game")
root.geometry("800x800")
root.configure(bg="black")

space = Label(root, text="                                             ", bg="black")
space.grid(row=0, column=0, rowspan=3)

heading = Label(root, text="Tic-Tac-Toe", bg="black", fg="white")
heading.configure(font = ("bold", 30))
heading.grid(row=0, column=1, columnspan=3)


XPoints = []
OPoints = []

def checkwin():
    global turnlabel
    global XPoints
    global OPoints
    
    def exist(l1, l2):
        for l in l1:
            if l not in l2:
                return False
        return True
    
    winning = [
        [(1, 1), (1, 2), (1, 3)],
        [(2, 1), (2, 2), (2, 3)],
        [(3, 1), (3, 2), (3, 3)],
        [(1, 1), (2, 1), (3, 1)],
        [(1, 2), (2, 2), (3, 2)],
        [(1, 3), (2, 3), (3, 3)],
        [(1, 1), (2, 2), (3, 3)],
        [(3, 1), (2, 2), (1, 3)]
    ]
    for win in winning:
        if exist(win, XPoints):
            turnlabel.configure(text="X WINS")
            for point in XOPoints:
                if (point.x, point.y) in win:
                    point.button.configure(bg="#90EE90")
            disable()
            break
        elif exist(win, OPoints):
            turnlabel.configure(text="O WINS")
            for point in XOPoints:
                if (point.x, point.y) in win:
                    point.button.configure(bg="#90EE90")
            disable()
            break
        elif len(XPoints)+len(OPoints)==9:
            turnlabel.configure(text="DRAW")
            for point in XOPoints:
                point.button.configure(bg="#FFCCCB")
            disable()
            break
        
def disable():
    global XOPoints
    for point in XOPoints:
        point.button.configure(state=DISABLED)
        
def restart():
    global XOPoints
    global XPoints
    global OPoints
    global current_chr
    
    XPoints = []
    OPoints = []
    current_chr = "X"
    turn = current_chr + " 's Turn" 
    turnlabel.configure(text = turn)
    
    
    
    for point in XOPoints:
        point.button.configure(state=NORMAL, bg="white")
        point.reset()

        
        

class XOPoint:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.value = None
        self.button = Button(root, text = "", width=10, height=5, command=self.set, font=("bold", 20), border=5)
        self.button.grid(row=x, column=y)
        
    def set(self):
        global current_chr
        global turnlabel
        global MAT
        if not self.value:
            self.button.configure(text=current_chr, bg='snow', fg='black')
            self.value = current_chr 
            if current_chr == "X":
                XPoints.append((self.x, self.y))
                MAT[self.x-1][self.y-1]="x"
                print(MAT)
                current_chr = "O"
            elif current_chr == "O":
                OPoints.append((self.x, self.y))
                MAT[self.x-1][self.y-1]="o"
                print(MAT)
                current_chr = "X"
        turn = current_chr + " 's Turn" 
        turnlabel.configure(text=turn)
        
        checkwin()
           
    def reset(self):
        self.button.configure(text = "")
        self.value = None
XOPoints=[]
MAT = [["b", "b", "b"],
      ["b", "b", "b"],
      ["b", "b", "b"]]
for x in range(1, 4):
    for y in range(1, 4):
        obj = XOPoint(x, y)
        XOPoints.append(obj)
current_chr = "X"
turn = current_chr + " 's Turn" 
turnlabel = Label(root, text=turn, bg="black", fg="white")
turnlabel.configure(font = ("bold", 30))
turnlabel.grid(row=5, column=1, columnspan=3)

playagain = Button(root, text="PLAY AGAIN", border=10, font=("bold", 20), command=restart)
playagain.grid(row=6, column=1, columnspan=3)

root.mainloop()
```
