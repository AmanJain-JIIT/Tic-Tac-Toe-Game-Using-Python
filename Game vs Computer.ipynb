```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("tic-tac-toe-endgame.csv")
data.head()
```

```python
df = data[data["V10"]=="negative"]
Data = df.drop(columns="V10")
```

```python
Data = Data.to_numpy()
```

```python
Data = np.reshape(Data, (Data.shape[0], 3, 3))
```

```python
inputmat = [['x', 'b', 'b'], ['b', 'b', 'b'], ['b', 'b', 'b']]
```

```python
def compare(mat):
    maxsim = Data[0]
    maxscore = 0
    for D in Data:
        simscore = 0
        for i in range(3):
            for j in range(3):
                if D[i][j]==mat[i][j]:
                    simscore+=1
        if simscore>maxscore:
            maxscore = simscore
            maxsim = D
    return maxsim
        
        
```

```python
def positions(M):
    L = []
    for i in range(3):
        for j in range(3):
            if M[i][j]=="o":
                L.append((i+1, j+1))
    return L
```

```python
from tkinter import *
from PIL import Image, ImageTk
import time

root = Tk()
root.title("Tic-Tac-Toe Game")
root.geometry("800x800")
root.configure(bg="black")

space = Label(root, text="                                             ", bg="black")
space.grid(row=0, column=0, rowspan=3)

heading = Label(root, text="Tic-Tac-Toe", bg="black", fg="white")
heading.configure(font = ("bold", 30))
heading.grid(row=0, column=1, columnspan=3)


XPoints = []
OPoints = []

def checkwin():
    global XPoints
    global OPoints
    global winlabel
    
    def exist(l1, l2):
        for l in l1:
            if l not in l2:
                return False
        return True
    
    winning = [
        [(1, 1), (1, 2), (1, 3)],
        [(2, 1), (2, 2), (2, 3)],
        [(3, 1), (3, 2), (3, 3)],
        [(1, 1), (2, 1), (3, 1)],
        [(1, 2), (2, 2), (3, 2)],
        [(1, 3), (2, 3), (3, 3)],
        [(1, 1), (2, 2), (3, 3)],
        [(3, 1), (2, 2), (1, 3)]
    ]
    for win in winning:
        if exist(win, XPoints):
            for point in XOPoints:
                if (point.x, point.y) in win:
                    winlabel.configure(text="YOU WON")
                    point.button.configure(bg="#90EE90")
            disable()
            break
        elif exist(win, OPoints):
            for point in XOPoints:
                if (point.x, point.y) in win:
                    winlabel.configure(text="YOU LOST")
                    point.button.configure(bg="#90EE90")
            disable()
            break
        elif len(XPoints)+len(OPoints)==9:
            for point in XOPoints:
                winlabel.configure(text="DRAW")
                point.button.configure(bg="#FFCCCB")
            disable()
            break
def Xwins():
    global XPoints
    
    winning = [
        [(1, 1), (1, 2), (1, 3)],
        [(2, 1), (2, 2), (2, 3)],
        [(3, 1), (3, 2), (3, 3)],
        [(1, 1), (2, 1), (3, 1)],
        [(1, 2), (2, 2), (3, 2)],
        [(1, 3), (2, 3), (3, 3)],
        [(1, 1), (2, 2), (3, 3)],
        [(3, 1), (2, 2), (1, 3)]
    ]
    res = []
    for win in winning:
        L = []
        for p in win:
            if p not in XPoints:
                L.append(p)
        res.append(L)
    for r in res:
        if len(r)==1:
            return r
    return 0
    
        
def disable():
    global XOPoints
    for point in XOPoints:
        point.button.configure(state=DISABLED)
        
def restart():
    global XOPoints
    global XPoints
    global OPoints
    global current_chr
    
    XPoints = []
    OPoints = []
    
    
    for point in XOPoints:
        point.button.configure(state=NORMAL, bg="white")
        point.reset()

        
def AIPlay():
    global MAT
    global XOPoints
    
    res = Xwins()
    
    if res!=0:
        for point in XOPoints:
            if point.x == res[0][0] and point.y==res[0][1] and point.value==None:
                point.button.configure(text="O", bg='snow', fg='black')
                point.value = "O"
                OPoints.append((point.x, point.y))
                MAT[point.x-1][point.y-1]="o"
                checkwin()
                return
        
    M = compare(MAT)
    L = positions(M)
    check = 0
    for point in XOPoints:
        if (point.x, point.y) in L and point.value==None:
            check=1
            point.button.configure(text="O", bg='snow', fg='black')
            point.value = "O"
            OPoints.append((point.x, point.y))
            MAT[point.x-1][point.y-1]="o"
            break
    if check==0:
        for point in XOPoints:
            if point.value==None:
                point.button.configure(text="O", bg='snow', fg='black')
                point.value = current_chr 
                OPoints.append((point.x, point.y))
                MAT[point.x-1][point.y-1]="o"
                break
    checkwin()
                
            
        
        

class XOPoint:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.value = None
        self.button = Button(root, text = "", width=10, height=5, command=self.set, font=("bold", 20), border=5)
        self.button.grid(row=x, column=y)
        
    def set(self):
        global turnlabel
        global MAT
        if not self.value:
            self.button.configure(text="X", bg='snow', fg='black')
            self.value = "X"
            XPoints.append((self.x, self.y))
            MAT[self.x-1][self.y-1]="x"
            print(MAT)
            checkwin()
            AIPlay()
            
    def reset(self):
        self.button.configure(text = "")
        self.value = None
XOPoints=[]
MAT = [["b", "b", "b"],
      ["b", "b", "b"],
      ["b", "b", "b"]]
for x in range(1, 4):
    for y in range(1, 4):
        obj = XOPoint(x, y)
        XOPoints.append(obj)
        
winlabel = Label(root, text="", font=("bold", 30), bg="black", fg="white")
winlabel.grid(row=5, column=1, columnspan=3, pady=20)

playagain = Button(root, text="PLAY AGAIN", border=10, font=("bold", 20), command=restart)
playagain.grid(row=6, column=1, columnspan=3)

root.mainloop()
```
